```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
folder = "<INSERT FILE PATH>"
filepath = folder + r"\rainfall_scrapped_weather_total_gov.csv"

scrapped_df = pd.read_csv(filepath)
scrapped_df.head(10)
```

```python
# Observe that the invalid characters are "\x97"
scrapped_df.iloc[2]["Highest 30 Min Rainfall (mm)"]
```

```python
# Perform data cleaning by replacing the bad values with np.nan
df = scrapped_df.replace('\x97', np.nan)
```
